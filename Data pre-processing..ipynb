### What is data Preprocessing ?

- Data preprocessing is a data mining technique that involves transforming raw data into an understandable format. 
- Real-world data is often incomplete, inconsistent, and/or lacking in certain behaviors or trends, and is likely to contain many errors.
- Data preprocessing is a proven method of resolving such issues.

### Why is Data Preprecessing important ?

- Machine learning algorithm follows a rule called GIGO(Garbage In Garbage Out.
    #### What are the proporties of raw data.
    - Incomplete
    - Noisy
    - Inconsistent
    - Duplicate
    
  #### What are the proporties of quality Preprocessed data.
    - Accuracy
    - Completeness
    - Consistancy
    - Believablity
    - Interpretability

### What are the steps to be follow for data Preprocessing.
- Data cleaning - This steps includes filling the missing values, smoothing the 
    noise & dealing the inconsistencis in the data.
- Data integration
- Data reduction
- Data transformation
- Data discretization

### What after Data Preprocessing..?
- Once the data preprocessing is we go for Feature Engineering.

### What is Feature Engineering?
- Feature Engineering is the process of creating new features or modifing existing features to get better performance from our Machine learning model.
- For feature engineering having a good domain knowledge is a must, because before we modify the data we should know what will be effect of the change on the data it self.
- Without proper domain knowledge we can also be ending by hampering the important features of the data.

### What are Feature ?
- Feature are the attributes or proporties of the unit of the data on which the analysis or the prediction is to be done.

### Why to do Feature Engineering?
- Well defined data and with more senceable features make the models prediction more accurate.
- It will also reduce the processing time.
- As the model is working on well preprocessed and well defined features the computation power will also decrease hence the cost of processing the data will also go down.
- Without feature engineering we may end with poor accuracy, more computing power and more processing time.

### What are the steps to be follow for Feature Engineering.
- Checking how the current features are working with the model.
- Brainstorming or testing the efficiency of the features. 
- Deciding what features to be created or modified.
- Creating new freatures or modifing old features.
- Checking how the new or modified features are working with the model.
- Improving the features of needed.
- It is not a one time done wala process, We need to repeat this all the steps again and again to get the best possible results.
- One thing to remember is that the feature engineering can only be done on the Preprocessed data, So we need to do data Preprocessing then we need to do feature engineering.

### Variable in Machine Learing.
- In general variables are the names given to a computer momory locations, which can be used to store values.
- But in Machine learning variable is any characteristic, number or quantity that can be measured or counted. 
- For example in the above dataset you can see the "Train Boarding Time" is the Feature and the value it contains are the variables.
- In the above dataset we can see the variables are the time.
- When we have to do operations on the variables we first need to know what is the type of the variable.

### Types of the variable.
- Numerical - Numbers which should be treated as they usually are in mathematics. For example, age and weight would be considered numerical variables
    - Discrete - A discrete variable is a variable whose value is obtained by counting. It is mostly integer nos. and the ex. of the features can be age, income etc.
    - Continuous - A continuous variable is a variable whose value is obtained by measuring. It is mostly floating type. and the ex. of the features can be percentage, weight etc.
- Categorical - A categorical variable is a variable whose values take on the value of labels. 
    - Ordinal - Ordinal variables have two are more categories that can be ordered or ranked.
    - Nominal - Nominal variables describe categories that do not have a specific order to them.
- Date and time - Dates and times are rich sources of information that can be used with machine learning models. 
- Mixed - These are random variables that are neither discrete nor continuous, but are a mixture of both.

### How to deal with different types of variables.
- Numerical variable are mostly used for performing arithimetical operations.
- Catagorical variables are mostly encoded by using One Hot encoding.

### Limitation of One Hot Encoding

- This method only works if the Categorical data is Nominal variable.
- Till now the above method looks cool but assume if we have a very huge data set with a catagorical data with 1000s of classes then it the model will need to have 1000s of dummy variable and 1000s of columns. 
- In such cases we need to take the most frequent classes and let say 80 % of data belongs to only 20 % of classes then we will make only those 20 dummy colums and one extra column in which rest of the 20% of the classes will be encoded.

### What is Ordinal Variables?

### What is Nominal Variables?
- Nominal variable is same like ordinal  variable but there is nothing that indicates an intrinsic order of the labels and in principle.

 - All labels are equal no order.

- No mathematical value

- Ex. Colour: Red, Green, White, Black

### What is Feature Scaling?
- Feature Scaling is a method to scale numeric features in the same scale or range (like:-1 to 1,  0 to 1).

- This is the last step involved in Data Preprocessing and before ML model training.

- It is also called as data normalization.

- We apply Feature Scaling on independent variables.

- We fit feature scaling with train data and transform on train and test data.

### Why Feature Scaling?
- The scale of raw features is different according to its units.

- Machine Learning algorithms cant understand features units, understand only numbers.

- Ex: If hight 140cm and 8.2feet

- ML Algorithms understand numbers then 140 > 8.2


### Which ML Algorithms Required Feature Scaling?
#### Those Algorithms Calculate Distance
- K-Nearest Neighbors (KNN)

- K-Means

- Support Vector Machine (SVM)

- Principal Component Analysis(PCA)

- Linear Discriminant Analysis

#### Gradient Descent Based Algorithms
- Linear Regression,

- Logistic Regression

- Neural Network

#### Tree Based Algorithms not required Feature scaling
- Decision Tree, Random Forest, XGBoost

#### Types of Feature Scaling
- 1) Min Max Scaler

- 2) Standard Scaler

- 3) Max Abs Scaler

- 4) Robust Scaler

- 5) Quantile Transformer Scaler

- 6) Power Transformer Scaler

- 7) Unit Vector Scaler


### Standardization vs Normalization Explain in Detail
#### What is Standardization?
- Standardization rescale the feature such as mean() = 0 and standard deviation () = 1.

- The result of standardization is Z called as Z-score normalization.

- If data follow a normal distribution (gaussian distribution).

- If the original distribution is normal, then the standardized distribution will be normal.

- If the original distribution is skewed, then the standardized distribution of the variable will also be skewed.

#### What is Normalization?
- Normalization rescale the feature in fixed range between 0 to 1.

- Normalization also called as Min-Max Scaling.

- If data doesnt follow normal distribution (Gaussian distribution).

#### Standardization  vs Normalization
- There is no any thumb rule to use Standardization or Normalization for special ML algo.

- But mostly Standardization use for clustering analyses, Principal Component Analysis(PCA).

- Normalization prefers for Image processing because of pixel intensity between 0 to 255, neural network algorithm requires data in scale 0-1, K-Nearest Neighbors.
